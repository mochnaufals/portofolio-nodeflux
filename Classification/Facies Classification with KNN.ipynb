```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Data Preprocessing

## Import Data

```python
df_A_raw = pd.read_csv('A.csv')
df_A_raw
```

```python
df_A_raw.describe().transpose()
```

```python
df_B_raw = pd.read_csv('B.csv')
df_B_raw
```

```python
df_B_raw.describe().transpose()
```

## Data Cleaning

```python
print(df_A_raw.isnull().values.any())
print(df_B_raw.isnull().values.any())
```

```python
df_A = df_A_raw.drop(labels=['RSHA', 'SP'], axis=1).dropna()
df_A
```

```python
df_B = df_B_raw.drop(labels=['RSHA', 'SP'], axis=1).dropna()
df_B
```

## Combine Data from A and B

```python
df = pd.concat([df_A, df_B])
df.reset_index(inplace=True, drop=True)
df
```

## Check Labels Count

```python
# Jumlah data tiap fasies
for i in np.unique(df['FACIES']):
    print(f'Fasies {i}: {sum(df["FACIES"] == i)}')
```

# 2. Feature Selection

## Spearman Correlation Matrix

```python
# Facies tidak dimasukkan karena bukan data numerik/ordinal
plt.figure(figsize=(12,12))
plt.suptitle('Spearman Correlation Matrix', fontsize=16, y=0.92)
sns.heatmap(df.drop(['FACIES'], axis=1).corr('spearman'), annot=True, vmin=-1, vmax=1, cmap='seismic_r')
plt.show()
```

```python
# Kolom yang didrop:
# RMED: resistivity medium, karena nilai Spearman = 1 dengan RDEP,
#       cukup pakai RDEP saja
# BS: bit size, karena bit size itu ditentukan oleh engineer di lapangan,
#     kurang merepresentasikan kondisi fasies sebenarnya, dan sudah diwakili
#     sebenarnya oleh ROP

SELECTED_FEATURES = [
    'DEPTH_MD',
    'CALI',
    # 'RMED',
    'RDEP',
    'RHOB',
    'GR',
    'NPHI',
    'PEF',
    'DTC',
    # 'BS',
    'ROP',
    'DRHO'
]
LABEL = 'FACIES'
```

# 3. Modelling

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df[SELECTED_FEATURES],
    df[LABEL],
    test_size=0.2,
    shuffle=True,
    random_state=7
)
```

## Data Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(X_train)

X_train_scaled = sc.transform(X_train)
X_test_scaled = sc.transform(X_test)
```

## Hyperparameter Tuning (with Validation Curve)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import validation_curve

model = KNeighborsClassifier(n_neighbors=3, weights='uniform')

param_range = np.arange(3, 20, 2)
train_scores, valid_scores = validation_curve(model,
                                              X_train_scaled, y_train, 
                                              param_name='n_neighbors',
                                              param_range=param_range,
                                              cv=4,
                                              scoring='balanced_accuracy')
```

```python
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
valid_scores_mean = np.mean(valid_scores, axis=1)
valid_scores_std = np.std(valid_scores, axis=1)
```

```python
train_scores
```

```python
plt.figure(figsize=(10,5))
plt.suptitle('Validation Curve', fontsize=14, y=0.94)
plt.plot(param_range, train_scores_mean, label='Training Score', color='red')
plt.fill_between(param_range, 
                 train_scores_mean - train_scores_std, 
                 train_scores_mean + train_scores_std, color='red', alpha=0.1)
plt.plot(param_range,
             valid_scores_mean,
             label='Validation Score',
             color='blue')
plt.fill_between(param_range, 
                 valid_scores_mean - valid_scores_std, 
                 valid_scores_mean + valid_scores_std, color='blue', alpha=0.1)
plt.legend()
plt.ylabel('Balanced Accuracy')
plt.xlabel('Neighbors')
plt.xlim(0, 12)
plt.ylim(0.8,1)
plt.grid()
```

```python
best_model = KNeighborsClassifier(n_neighbors=6, weights='uniform')
best_model.fit(X_train_scaled, y_train)
```

# 4. Model Testing

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, balanced_accuracy_score, precision_score, recall_score
```

## Metrics Score

```python
y_pred = best_model.predict(X_test_scaled)
print('Balanced Accuracy:', balanced_accuracy_score(y_test, y_pred))
print('F1 Score:', f1_score(y_test, y_pred, average='weighted'))
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, y_pred)
cm

ConfusionMatrixDisplay(cm).plot()
```

# 5. Model Inspection

```python
from sklearn.inspection import permutation_importance

r = permutation_importance(best_model, 
                           X_test_scaled, 
                           y_test,
                           n_repeats=30)

for i in r.importances_mean.argsort()[::-1]:
  print(f'{X_train.columns[i]:<8}\t'
        f'{r.importances_mean[i]:.3f}'
        f' +/- {r.importances_std[i]:.3f}')
```

```python
tree_feature_importances = r.importances_mean
sorted_idx = tree_feature_importances.argsort()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
fig.suptitle('Permutation Feature Importance', fontsize=14, y=0.94)
y_ticks = np.arange(0, len(X_train.columns))
ax.barh(y_ticks, tree_feature_importances[sorted_idx])
ax.set_yticks(y_ticks)
ax.set_yticklabels(X_train.columns[sorted_idx])
ax.set_xlabel('Tree Feature Importance Value')
plt.show()
```
