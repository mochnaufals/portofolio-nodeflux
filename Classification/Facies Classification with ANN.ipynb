```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Data Preprocessing

## Import Data

```python
df_A_raw = pd.read_csv('A.csv')
df_A_raw
```

```python
df_A_raw.describe().transpose()
```

```python
df_B_raw = pd.read_csv('B.csv')
df_B_raw
```

```python
df_B_raw.describe().transpose()
```

## Data Cleaning

```python
print(df_A_raw.isnull().values.any())
print(df_B_raw.isnull().values.any())
```

```python
df_A = df_A_raw.drop(labels=['RSHA', 'SP'], axis=1).dropna()
df_A
```

```python
df_B = df_B_raw.drop(labels=['RSHA', 'SP'], axis=1).dropna()
df_B
```

## Combine Data from A and B

```python
df = pd.concat([df_A, df_B])
df.reset_index(inplace=True, drop=True)
df
```

## Check Labels Count

```python
# Jumlah data tiap fasies
for i in np.unique(df['FACIES']):
    print(f'Fasies {i}: {sum(df["FACIES"] == i)}')
```

# 2. Feature Selection

## Spearman Correlation Matrix

```python
# Facies tidak dimasukkan karena bukan data numerik/ordinal
plt.figure(figsize=(12,12))
plt.suptitle('Spearman Correlation Matrix', fontsize=16, y=0.92)
sns.heatmap(df.drop(['FACIES'], axis=1).corr('spearman'), annot=True, vmin=-1, vmax=1, cmap='seismic_r')
plt.show()
```

```python
# Kolom yang didrop:
# RMED: resistivity medium, karena nilai Spearman = 1 dengan RDEP,
#       cukup pakai RDEP saja
# BS: bit size, karena bit size itu ditentukan oleh engineer di lapangan,
#     kurang merepresentasikan kondisi fasies sebenarnya, dan sudah diwakili
#     sebenarnya oleh ROP

SELECTED_FEATURES = [
    'DEPTH_MD',
    'CALI',
    # 'RMED',
    'RDEP',
    'RHOB',
    'GR',
    'NPHI',
    'PEF',
    'DTC',
    # 'BS',
    'ROP',
    'DRHO'
]
LABEL = 'FACIES'
```

# 3. Modelling

## Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df[SELECTED_FEATURES],
    df[LABEL],
    test_size=0.2,
    shuffle=True,
    random_state=7
)
```

## Data Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc.fit(X_train)

X_train_scaled = sc.transform(X_train)
X_test_scaled = sc.transform(X_test)
```

## Hyperparameter Tuning (with GridSearchCV)

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'hidden_layer_sizes': [
        # (10, 10, 10),
        # (20, 20, 20),
        # (30, 30, 30),
        # (40, 40, 40),
        (50, 50, 50),
        # (60, 60, 60),
        (70, 70, 70),
        # (80, 80, 80),
        (90, 90, 90),
        # (100, 100, 100)
    ],
    # 'solver': ['adam', 'lbfgs'],
    'learning_rate_init': [0.01, 0.02]
}

model = MLPClassifier(
    hidden_layer_sizes=(10, 10, 10),
    activation='relu',
    solver='adam',
    learning_rate_init=0.01,
    random_state=7
)

search = GridSearchCV(model, param_grid, cv=5)
search.fit(X_train_scaled, y_train)
```

```python
search.best_params_
```

```python
search.best_score_
```

```python
best_model = search.best_estimator_
best_model.fit(X_train_scaled, y_train)
```

# 4. Model Testing

```python
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, balanced_accuracy_score, precision_score, recall_score
```

## Metrics Score

```python
y_pred = best_model.predict(X_test_scaled)
print('Balanced Accuracy:', balanced_accuracy_score(y_test, y_pred))
print('F1 Score:', f1_score(y_test, y_pred, average='weighted'))
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, y_pred)
cm

ConfusionMatrixDisplay(cm).plot()
```

# 5. Model Inspection

```python
from sklearn.inspection import permutation_importance

r = permutation_importance(best_model, 
                           X_test_scaled, 
                           y_test,
                           n_repeats=30)

for i in r.importances_mean.argsort()[::-1]:
  print(f'{X_train.columns[i]:<8}\t'
        f'{r.importances_mean[i]:.3f}'
        f' +/- {r.importances_std[i]:.3f}')
```

```python
tree_feature_importances = r.importances_mean
sorted_idx = tree_feature_importances.argsort()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
fig.suptitle('Permutation Feature Importance', fontsize=14, y=0.94)
y_ticks = np.arange(0, len(X_train.columns))
ax.barh(y_ticks, tree_feature_importances[sorted_idx])
ax.set_yticks(y_ticks)
ax.set_yticklabels(X_train.columns[sorted_idx])
ax.set_xlabel('Tree Feature Importance Value')
plt.show()
```
