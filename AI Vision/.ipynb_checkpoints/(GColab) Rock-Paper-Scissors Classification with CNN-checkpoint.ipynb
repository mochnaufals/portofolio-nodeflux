# Inisiasi

```
# Import kelas dan fungsi yang dibutuhkan

# TensorFlow
import tensorflow as tf
from tensorflow import keras
from keras import layers
from keras.optimizers import adamax_v2 as Adamax
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import TensorBoard
from time import time

# Archive dan direktori
import zipfile,os

# Numpy array
import numpy as np

# Files dari google colab
from google.colab import files

# Library Keras
from keras.preprocessing import image

# Plot gambar
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Split folder
!pip install split_folders
import splitfolders as split_folders
```

```
# Import dataset dari blob dicoding dengan wget

!wget --no-check-certificate \
  https://github.com/dicodingacademy/assets/releases/download/release/rockpaperscissors.zip \
  -O /tmp/dataset.zip
```

```
# Ekstrak dataset dari zip

datazip = '/tmp/dataset.zip'
refzip = zipfile.ZipFile(datazip, 'r')

refzip.extractall('/tmp')
refzip.close()
os.remove('/tmp/rockpaperscissors/rps-cv-images/README_rpc-cv-images.txt')
```

```
# Membagi dataset menjadi training set dan validation set

dir_dataset = '/tmp/rockpaperscissors/rps-cv-images'

split_folders.ratio(dir_dataset, output='/tmp/databaru', ratio=(0.6, 0.4)) # Validation set sudah 40 persen

dir_train = os.path.join('/tmp/databaru', 'train')
dir_valid = os.path.join('/tmp/databaru', 'val')
```

# Augmentasi Data

```
# Membuat obyek ImageDataGenerator train dan validation

datagen_train = ImageDataGenerator(
    rescale=1./255,
    rotation_range=45,
    vertical_flip=True,
    horizontal_flip=True,
    shear_range=0.1,
    fill_mode='nearest',
    brightness_range=[0.5, 1]
)

datagen_valid = ImageDataGenerator(
    rescale=1./255,
    rotation_range=45,
    vertical_flip=True,
    horizontal_flip=True,
    shear_range=0.1,
    fill_mode='nearest',
    brightness_range=[0.5, 1]
)
```

```
# Membuat obyek generatornya

generator_train = datagen_train.flow_from_directory(
    dir_train,
    target_size=(150, 150),
    batch_size=16,
    class_mode='categorical'
)

generator_valid = datagen_valid.flow_from_directory(
    dir_valid,
    target_size=(150, 150),
    batch_size=16,
    class_mode='categorical'
)

# Catatan tambahan ############################################################
# Mohon maaf jika jumlah validation set nya 876 bukan 874                     #
# Saya sudah coba hitung ternyata pembagiannya menghasilkan bilangan desimal  #
# Sehingga ada perbedaan pembulatan                                           #
# Namun validation set sudah saya atur sebanyak 40 persen                     #
###############################################################################
```

# Melatih Model

```
# Membuat model Sequential

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')
])
```

```
model.compile(
    loss='categorical_crossentropy',
    optimizer=tf.optimizers.Adamax(learning_rate=0.001),
    metrics=['accuracy']
)
```

```
model.fit(
    generator_train,
    steps_per_epoch=25,
    epochs=50,
    validation_data=generator_valid,
    validation_steps=5,
    verbose=2,
    callbacks=[TensorBoard(log_dir='/tmp/logtensor/{}'.format(time()))],
)
```

# Pengujian Model
